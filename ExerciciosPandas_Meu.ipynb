**Bora praticar!**
  
1) Utilizando o DataFrame importado (alunos3.csv) calcule a média das provas em uma nova coluna chamada (Media_provas)

```python
df = pd.read_csv('dados/alunos3.csv', sep= ';', decimal= ',')
df
```

```python
df['Media_provas'] = (df['Prova_1'] + df['Prova_2'] + df['Prova_3'] + df['Prova_4']) / 4
df
```

2) Quem foram os alunos que obtiveram a maior e a menor média

```python
ordenado = df.sort_values(by=['Media_provas'], ascending=False).reset_index(drop='True')
maior = df['Nome'][0]
menor = df['Nome'][ordenado.shape[0]-1]
print(maior, menor)
```

3) Agora una este dataframe com o `cadastro_alunos.xlsx`

```python
df_2 = pd.read_excel('dados/cadastro_alunos.xlsx')
df_2

tabela = df.merge(df_2, how = 'left', on = 'RA')
tabela
```

4) Qual a média entre as Media_provas dentro do público feminino? e masculino?

```python
df_medias = tabela.groupby('Sexo').agg({'Prova_1' : 'mean',
                                             'Prova_2' : 'mean',
                                             'Prova_3' : 'mean',
                                             'Prova_4' : 'mean',
                                             'Media_provas' : 'mean'})
df_medias
```

5) Qual a média de idade das pessoas que obtiveram Media_provas maior ou igual a 7?

```python
tabela['Tratamento'] = tabela['Media_provas'].apply(lambda x: 1 if x >= 7 else 0)
tabela

df_q5 = tabela.groupby('Tratamento').agg({'Idade' : 'mean'}).reset_index()
df_q5.columns = ['Criterio', 'Media_idade']

df_q5['Media_idade'][1]
```

6) Qual das cidades possui o maior média de Media_provas? E qual é este valor?

```python
df_cidades = tabela.groupby('Cidade').agg({'Media_provas' : 'max'}).reset_index()
df_cidades.head(1)
```
